# **Feature Engineering & Limpieza**

```python
df_model = df_valid.copy()

# ============================================================================
# 1. SELECCIONAR FEATURES RELEVANTES Y ELIMINAR COLUMNAS INÚTILES
# ============================================================================

print("\n1. Seleccionando features...")

# Eliminar columnas con >80% nulos o no útiles para predicción
drop_columns = [
    'recommendations',
    'tagline',
    'backdrop_path',
    'poster_path',
    'id',
    'title'
]

df_model = df_model.drop(columns=drop_columns, errors='ignore')
print(f"Eliminadas columnas: {drop_columns}")
```

```python
# ============================================================================
# 2. FEATURE ENGINEERING: LENGUAJE
# ============================================================================

print("\n2. Feature Engineering: original_language...")

# Agrupar idiomas: inglés vs otros
df_model['is_english'] = (df_model['original_language'] == 'en').astype(int)

# Top idiomas
top_languages = df_model['original_language'].value_counts().head(10).index.tolist()
df_model['language_group'] = df_model['original_language'].apply(
    lambda x: x if x in top_languages else 'other'
)

print(f"Creados: is_english, language_group")
print(f"  Idiomas principales: {top_languages}")
```

```python
# ============================================================================
# 3. FEATURE ENGINEERING: FECHAS
# ============================================================================

print("\n3. Feature Engineering: release_date...")

df_model['release_date'] = pd.to_datetime(df_model['release_date'], errors='coerce')
df_model['year'] = df_model['release_date'].dt.year
df_model['month'] = df_model['release_date'].dt.month
df_model['quarter'] = df_model['release_date'].dt.quarter
df_model['day_of_week'] = df_model['release_date'].dt.dayofweek

# Categorías de temporada de cine
def get_season(month):
    if pd.isna(month):
        return 'unknown'
    if month in [1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'fall'

df_model['season'] = df_model['month'].apply(get_season)

# Películas recientes (últimos 5 años)
max_year = df_model['year'].max()
df_model['is_recent'] = (df_model['year'] >= max_year - 5).astype(int)

print(f"Creados: year, month, quarter, day_of_week, season, is_recent")
print(f"  Rango de años: {df_model['year'].min()} - {df_model['year'].max()}")
```

```python
# ============================================================================
# 4. FEATURE ENGINEERING: GÉNEROS (Multi-label)
# ============================================================================

print("\n4. Feature Engineering: géneros...")

def parse_genres_delimited(genre_str):
    """Parsear géneros separados por guiones"""
    if pd.isna(genre_str) or genre_str == '':
        return []

    # Dividir por guión y limpiar espacios
    genres = [g.strip().lower() for g in str(genre_str).split('-')]
    # Filtrar strings vacíos
    genres = [g for g in genres if g]
    return genres

# Aplicar función
df_model['genres_list'] = df_model['genres'].apply(parse_genres_delimited)

# Verificar cuántos géneros se parsearon
parsed_count = df_model['genres_list'].apply(len).sum()
print(f"✓ Géneros parseados exitosamente")
print(f"  Total de géneros extraídos: {parsed_count}")
print(f"  Películas sin géneros: {(df_model['genres_list'].apply(len) == 0).sum()}")

# Ver ejemplos
print("\nEjemplos de géneros parseados:")
for i in range(min(5, len(df_model))):
    if df_model['genres_list'].iloc[i]:
        print(f"  Película {i}: {df_model['genres_list'].iloc[i]}")

# Top géneros
all_genres = []
for genres_list in df_model['genres_list']:
    all_genres.extend(genres_list)

from collections import Counter
genre_counts = Counter(all_genres)
top_genres = [g for g, count in genre_counts.most_common(15)]  # Top 15 en lugar de 10

print(f"\nTop 15 géneros encontrados:")
for genre, count in genre_counts.most_common(15):
    pct = count/len(df_model)*100
    print(f"  - {genre.capitalize():20s}: {count:5d} películas ({pct:5.2f}%)")

# Crear features binarias para top géneros
for genre in top_genres:
    df_model[f'genre_{genre}'] = df_model['genres_list'].apply(
        lambda x: 1 if genre in x else 0
    )

print(f"\nCreadas {len(top_genres)} features binarias de géneros")

# Número de géneros por película
df_model['num_genres'] = df_model['genres_list'].apply(len)
print(f"Creado: num_genres")
print(f"  - Media de géneros por película: {df_model['num_genres'].mean():.2f}")
print(f"  - Rango: {df_model['num_genres'].min()} a {df_model['num_genres'].max()}")

# Verificar distribución
print(f"\nDistribución de num_genres:")
print(df_model['num_genres'].value_counts().sort_index())
```

```python
# ============================================================================
# 5. FEATURE ENGINEERING: BUDGET
# ============================================================================

print("\n5. Feature Engineering: budget...")

df_model['has_budget'] = (df_model['budget'] > 0).astype(int)
df_model['budget_log'] = np.log1p(df_model['budget'])

def categorize_budget(budget):
    if budget == 0:
        return 'no_budget'
    elif budget < 1_000_000:
        return 'low'
    elif budget < 10_000_000:
        return 'medium'
    elif budget < 50_000_000:
        return 'high'
    else:
        return 'very_high'

df_model['budget_category'] = df_model['budget'].apply(categorize_budget)

print(f"Creados: has_budget, budget_log, budget_category")
print(f"  Distribución de budget_category:")
print(df_model['budget_category'].value_counts())
```

```python
# ============================================================================
# 6. FEATURE ENGINEERING: STATUS
# ============================================================================

print("\n6. Feature Engineering: status...")

df_model['is_released'] = (df_model['status'] == 'Released').astype(int)

print(f"Creado: is_released")
print(f"  Películas lanzadas: {df_model['is_released'].sum()} ({df_model['is_released'].mean()*100:.2f}%)")
```

```python
# ============================================================================
# 7. TRATAMIENTO DE OUTLIERS EN VARIABLES NUMÉRICAS
# ============================================================================

print("\n7. Tratamiento de outliers...")

numeric_features = ['popularity', 'budget', 'runtime', 'vote_average', 'vote_count']

# Aplicar Winsorizing (capping) en lugar de eliminar
from scipy.stats import mstats

for feature in numeric_features:
    lower = df_model[feature].quantile(0.01)
    upper = df_model[feature].quantile(0.99)
    
    before_outliers = ((df_model[feature] < lower) | (df_model[feature] > upper)).sum()
    
    df_model[feature] = np.where(df_model[feature] < lower, lower, df_model[feature])
    df_model[feature] = np.where(df_model[feature] > upper, upper, df_model[feature])
    
    print(f"  {feature}: {before_outliers} outliers reemplazados (P1-P99 capping)")
```

```python
# ============================================================================
# 8. IMPUTACIÓN DE VALORES NULOS 
# ============================================================================

print("\n8. Imputación de valores nulos...")

# Runtime: mediana (es numérica)
if 'runtime' in df_model.columns:
    median_runtime = df_model['runtime'].median()
    null_runtime_before = df_model['runtime'].isnull().sum()
    df_model['runtime'].fillna(median_runtime, inplace=True)
    print(f"  runtime: {null_runtime_before} valores imputados con mediana = {median_runtime:.2f}")

# Overview: llenar con string vacío (es texto)
if 'overview' in df_model.columns:
    null_overview_before = df_model['overview'].isnull().sum()
    df_model['overview'].fillna('', inplace=True)
    print(f"  overview: {null_overview_before} valores imputados con string vacío")

# Verificar que no hay más nulos en features numéricas críticas
print(f"\nValores nulos restantes en features críticas:")
critical_features = ['popularity', 'budget', 'runtime', 'vote_average', 'vote_count']
for feat in critical_features:
    if feat in df_model.columns:
        null_count = df_model[feat].isnull().sum()
        print(f"  {feat}: {null_count}")
```

```python
# ============================================================================
# 9. CODIFICACIÓN DE VARIABLES CATEGÓRICAS
# ============================================================================

print("\n9. Codificación de variables categóricas...")

# One-hot encoding para language_group
language_dummies = pd.get_dummies(df_model['language_group'], prefix='lang')
df_model = pd.concat([df_model, language_dummies], axis=1)

# One-hot encoding para season
season_dummies = pd.get_dummies(df_model['season'], prefix='season')
df_model = pd.concat([df_model, season_dummies], axis=1)

# One-hot encoding para budget_category
budget_dummies = pd.get_dummies(df_model['budget_category'], prefix='budget_cat')
df_model = pd.concat([df_model, budget_dummies], axis=1)

print(f"Creadas dummies para: language_group, season, budget_category")
```

```python
# ============================================================================
# 10. SELECCIONAR FEATURES FINALES PARA MODELO
# ============================================================================

print("\n10. Seleccionando features finales...")

# Features numéricas
numeric_features_final = [
    'popularity', 'budget', 'runtime', 'vote_average', 'vote_count',
    'budget_log', 'num_genres', 'has_budget', 'is_english', 'is_recent', 
    'is_released', 'day_of_week', 'month', 'quarter'
]

# Features binarias de géneros
genre_features = [col for col in df_model.columns if col.startswith('genre_')]

# Features categóricas codificadas
language_features = [col for col in df_model.columns if col.startswith('lang_')]
season_features = [col for col in df_model.columns if col.startswith('season_')]
budget_cat_features = [col for col in df_model.columns if col.startswith('budget_cat_')]

# Compilar lista de features
X_features = (numeric_features_final + genre_features + 
              language_features + season_features + budget_cat_features)

print(f"Features seleccionadas: {len(X_features)}")
print(f"\n  Numéricas: {len(numeric_features_final)}")
print(f"  Géneros: {len(genre_features)}")
print(f"  Idiomas: {len(language_features)}")
print(f"  Estaciones: {len(season_features)}")
print(f"  Budget categorizado: {len(budget_cat_features)}")

print(f"\nLista de features:")
for i, feat in enumerate(X_features, 1):
    print(f"  {i}. {feat}")
```

```python
# ============================================================================
# 11. CREAR MATRIZ X Y VECTOR y
# ============================================================================

print("\n11. Creando matrices X y y...")

X = df_model[X_features].copy()
y = df_model['exito_comercial'].copy()

# Verificar valores nulos restantes
print(f"\nValores nulos en X:")
print(X.isnull().sum()[X.isnull().sum() > 0])

# Rellenar cualquier valor nulo restante
X.fillna(X.mean(numeric_only=True), inplace=True)

print(f"Matriz X: {X.shape}")
print(f"Vector y: {y.shape}")
print(f"\nDistribución de target:")
print(y.value_counts())
print(f"\nProporciones:")
print(y.value_counts(normalize=True))
```

```python
# ============================================================================
# 12. RESUMEN DE FEATURES
# ============================================================================

print("\n" + "="*80)
print("RESUMEN DE FEATURE ENGINEERING")
print("="*80)

print(f"""
FEATURES CREADAS:
   - Temporales: year, month, quarter, day_of_week, season, is_recent
   - Idioma: is_english, language_group (one-hot encoded)
   - Géneros: {len(genre_features)} features binarias + num_genres
   - Budget: has_budget, budget_log, budget_category (one-hot encoded)
   - Status: is_released
   
TRATAMIENTOS APLICADOS:
   - Outliers: Winsorizing P1-P99 en variables numéricas
   - Nulos: Mediana para runtime y overview
   - Codificación: One-hot para categóricas
   
DATASET FINAL:
   - Muestras: {X.shape[0]}
   - Features: {X.shape[1]}
   - Target: exito_comercial (0/1)
   - Balance: {(y==1).sum()/len(y)*100:.2f}% exitosas
   
PRÓXIMO PASO: Train-Test Split y Modelado
""")

# Guardar para siguiente paso
print("\nListo para Train-Test Split y modelado")
print(f"  X.shape = {X.shape}")
print(f"  y.shape = {y.shape}")
```

```python
# ============================================================================
# LIMPIEZA POST FEATURE ENGINEERING
# ============================================================================

print("="*80)
print("LIMPIEZA FINAL: Valores nulos en features temporales")
print("="*80)

# Llenar nulos en features temporales con la moda 
temporal_features = ['day_of_week', 'month', 'quarter']

for feat in temporal_features:
    if feat in X.columns:
        null_count = X[feat].isnull().sum()
        mode_val = X[feat].mode()[0] if len(X[feat].mode()) > 0 else 0
        X[feat].fillna(mode_val, inplace=True)
        print(f"  {feat}: {null_count} nulos imputados con moda = {mode_val}")

# Verificar que no quedan nulos
print(f"\nVerificación final - Valores nulos en X:")
print(X.isnull().sum().sum())
```

```python
# ============================================================================
# NORMALIZACIÓN/ESCALADO DE FEATURES
# ============================================================================

print("\n" + "="*80)
print("NORMALIZACIÓN DE FEATURES")
print("="*80)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convertir a DataFrame para mantener nombres de columnas
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns, index=X_test.index)

print(f"Features escalados con StandardScaler")
print(f"\nEstadísticas de X_train_scaled:")
print(f"  Media: {X_train_scaled.mean().mean():.6f} (≈ 0)")
print(f"  Desv std: {X_train_scaled.std().mean():.6f} (≈ 1)")
```

```python
# ============================================================================
# BASELINE MODEL: DUMMY CLASSIFIER 
# ============================================================================

print("\n" + "="*80)
print("BASELINE MODEL: DUMMY CLASSIFIER")
print("="*80)

# Baseline 1: Estrategia "most_frequent" (siempre predice clase mayoritaria)
baseline_mf = DummyClassifier(strategy='most_frequent', random_state=42)
baseline_mf.fit(X_train_scaled, y_train)
y_pred_mf = baseline_mf.predict(X_test_scaled)

# Obtener la clase mayoritaria
most_frequent_class = y_train.mode()[0]

print(f"\n1. Baseline - Most Frequent (siempre predice clase mayoritaria):")
print(f"   Predice siempre: {most_frequent_class} (no exitosa)")
print(f"   Accuracy: {accuracy_score(y_test, y_pred_mf):.4f}")
print(f"   Precision (exitosa): {precision_score(y_test, y_pred_mf, zero_division=0):.4f}")
print(f"   Recall (exitosa): {recall_score(y_test, y_pred_mf, zero_division=0):.4f}")
print(f"   F1-Score: {f1_score(y_test, y_pred_mf, zero_division=0):.4f}")

# Baseline 2: Estrategia "stratified" (predice aleatoriamente respetando proporciones)
baseline_strat = DummyClassifier(strategy='stratified', random_state=42)
baseline_strat.fit(X_train_scaled, y_train)
y_pred_strat = baseline_strat.predict(X_test_scaled)

print(f"\n2. Baseline - Stratified (predice aleatoriamente con proporciones):")
print(f"   Accuracy: {accuracy_score(y_test, y_pred_strat):.4f}")
print(f"   Precision (exitosa): {precision_score(y_test, y_pred_strat, zero_division=0):.4f}")
print(f"   Recall (exitosa): {recall_score(y_test, y_pred_strat, zero_division=0):.4f}")
print(f"   F1-Score: {f1_score(y_test, y_pred_strat, zero_division=0):.4f}")

# Baseline 3: Modelo simple con Logistic Regression (baseline numérico)
from sklearn.linear_model import LogisticRegression

baseline_lr = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
baseline_lr.fit(X_train_scaled, y_train)
y_pred_lr = baseline_lr.predict(X_test_scaled)
y_pred_lr_proba = baseline_lr.predict_proba(X_test_scaled)

print(f"\n3. Baseline - Logistic Regression Simple:")
print(f"   Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
print(f"   Precision (exitosa): {precision_score(y_test, y_pred_lr, zero_division=0):.4f}")
print(f"   Recall (exitosa): {recall_score(y_test, y_pred_lr, zero_division=0):.4f}")
print(f"   F1-Score: {f1_score(y_test, y_pred_lr, zero_division=0):.4f}")

# Matriz de confusión
cm = confusion_matrix(y_test, y_pred_lr)
print(f"\n   Matriz de confusión (LR):")
print(f"   [[TN={cm[0,0]:5d}  FP={cm[0,1]:5d}]")
print(f"    [FN={cm[1,0]:5d}  TP={cm[1,1]:5d}]]")
```

```python
# ============================================================================
# MODELO 1: RANDOM FOREST
# ============================================================================

results = {}

print("\n1. RANDOM FOREST")
print("-" * 80)

start_time = time.time()

rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=15,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

rf_model.fit(X_train_scaled, y_train)
y_pred_rf = rf_model.predict(X_test_scaled)
y_pred_rf_proba = rf_model.predict_proba(X_test_scaled)[:, 1]

rf_time = time.time() - start_time

print(f"Tiempo de entrenamiento: {rf_time:.2f}s")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_rf):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_rf):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_pred_rf):.4f}")

results['Random Forest'] = {
    'Accuracy': accuracy_score(y_test, y_pred_rf),
    'Precision': precision_score(y_test, y_pred_rf),
    'Recall': recall_score(y_test, y_pred_rf),
    'F1-Score': f1_score(y_test, y_pred_rf),
    'Time': rf_time,
    'y_pred': y_pred_rf,
    'y_proba': y_pred_rf_proba
}

# Feature importance
feature_importance_rf = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)

print(f"\nTop 10 features más importantes:")
print(feature_importance_rf.head(10).to_string(index=False))
```

```python
# ============================================================================
# MODELO 2: XGBOOST
# ============================================================================

print("\n\n2. XGBOOST")
print("-" * 80)

start_time = time.time()

xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    scale_pos_weight=(y_train == 0).sum() / (y_train == 1).sum(),  # Balance de clases
    n_jobs=-1,
    verbosity=0
)

xgb_model.fit(X_train_scaled, y_train)
y_pred_xgb = xgb_model.predict(X_test_scaled)
y_pred_xgb_proba = xgb_model.predict_proba(X_test_scaled)[:, 1]

xgb_time = time.time() - start_time

print(f"Tiempo de entrenamiento: {xgb_time:.2f}s")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_xgb):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_xgb):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_xgb):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_pred_xgb):.4f}")

results['XGBoost'] = {
    'Accuracy': accuracy_score(y_test, y_pred_xgb),
    'Precision': precision_score(y_test, y_pred_xgb),
    'Recall': recall_score(y_test, y_pred_xgb),
    'F1-Score': f1_score(y_test, y_pred_xgb),
    'Time': xgb_time,
    'y_pred': y_pred_xgb,
    'y_proba': y_pred_xgb_proba
}

# Feature importance
feature_importance_xgb = pd.DataFrame({
    'feature': X.columns,
    'importance': xgb_model.feature_importances_
}).sort_values('importance', ascending=False)

print(f"\nTop 10 features más importantes:")
print(feature_importance_xgb.head(10).to_string(index=False))
```

```python
# ============================================================================
# MODELO 3: RIDGE 
# ============================================================================

print("\n\n3. RIDGE")
print("-" * 80)

start_time = time.time()

# Ridge necesita normalización (ya tenemos con StandardScaler)
ridge_model = Ridge(alpha=1.0, random_state=42)
ridge_model.fit(X_train_scaled, y_train)
y_pred_ridge_proba = ridge_model.predict(X_test_scaled)

ridge_time = time.time() - start_time

# Convertir a clasificación con threshold 0.5
y_pred_ridge = (y_pred_ridge_proba >= 0.5).astype(int)

print(f"Tiempo de entrenamiento: {ridge_time:.2f}s")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_ridge):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_ridge):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_ridge):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_pred_ridge):.4f}")

results['Ridge'] = {
    'Accuracy': accuracy_score(y_test, y_pred_ridge),
    'Precision': precision_score(y_test, y_pred_ridge),
    'Recall': recall_score(y_test, y_pred_ridge),
    'F1-Score': f1_score(y_test, y_pred_ridge),
    'Time': ridge_time,
    'y_pred': y_pred_ridge,
    'y_proba': y_pred_ridge_proba
}

# Coefficients
coef_importance = pd.DataFrame({
    'feature': X.columns,
    'coefficient': np.abs(ridge_model.coef_)
}).sort_values('coefficient', ascending=False)

print(f"\nTop 10 features más importantes (por coeficiente):")
print(coef_importance.head(10).to_string(index=False))
```

```python
# ============================================================================
# MODELO 4: LASSO 
# ============================================================================

print("\n\n4. LASSO")
print("-" * 80)

start_time = time.time()

lasso_model = Lasso(alpha=0.01, max_iter=10000, random_state=42)
lasso_model.fit(X_train_scaled, y_train)
y_pred_lasso_proba = lasso_model.predict(X_test_scaled)

lasso_time = time.time() - start_time

# Convertir a clasificación con threshold 0.5
y_pred_lasso = (y_pred_lasso_proba >= 0.5).astype(int)

print(f"Tiempo de entrenamiento: {lasso_time:.2f}s")
print(f"Accuracy:  {accuracy_score(y_test, y_pred_lasso):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_lasso):.4f}")
print(f"Recall:    {recall_score(y_test, y_pred_lasso):.4f}")
print(f"F1-Score:  {f1_score(y_test, y_pred_lasso):.4f}")

results['Lasso'] = {
    'Accuracy': accuracy_score(y_test, y_pred_lasso),
    'Precision': precision_score(y_test, y_pred_lasso),
    'Recall': recall_score(y_test, y_pred_lasso),
    'F1-Score': f1_score(y_test, y_pred_lasso),
    'Time': lasso_time,
    'y_pred': y_pred_lasso,
    'y_proba': y_pred_lasso_proba
}

# Coefficients (solo las no-cero)
coef_lasso = pd.DataFrame({
    'feature': X.columns,
    'coefficient': np.abs(lasso_model.coef_)
})
coef_lasso = coef_lasso[coef_lasso['coefficient'] > 0].sort_values('coefficient', ascending=False)

print(f"\nFeatures seleccionadas por Lasso: {len(coef_lasso)} de {len(X.columns)}")
if len(coef_lasso) > 0:
    print(f"Top 10 features:")
    print(coef_lasso.head(10).to_string(index=False))
```
